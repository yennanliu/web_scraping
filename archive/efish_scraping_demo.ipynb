```python
# import library 
import requests
from bs4 import BeautifulSoup
import pandas as pd 
```

```python
# parse 
headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_13_6) AppleWebKit/537.36 (KHTML, like Gecko) Chorme/68.0.3440.106 Safari/537.36'}

URL = "https://efish.fa.gov.tw/efish/statistics/coraltradedatalist.htm"
data = {
    "calendarType": "tw",
    "dateStrl": "89.1.1",
    "dateStrt": "108.3.21",
    "yearl": "89",
    "yeart": "108",
    "monthl": "1",
    "montht": "3",
    "dayl": "1",
    "dayt": "21",
    "species": "-1",
    "spall": "-1",
    "orderby":"0"
    }
res = requests.post(URL, data=data, headers=headers)
#print(res.text)
soup = BeautifulSoup(res.text, 'html.parser')
```

```python
# digest and save to output 
output = []
divs = soup.findAll("table", {"class": "bk_white"})

for div in divs:
    row = ''
    rows = div.findAll('tr')
    #count =1 
    for row in rows:
        #print ('count :', count)
        #if(row.text.find("PHONE") > -1):
        print(row.text)
        #count = count + 1 
        output.append(row.text)
```
