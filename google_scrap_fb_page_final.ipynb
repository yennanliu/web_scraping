```python
# REF : http://stackoverflow.com/questions/4371655/extract-google-search-results

import sys # Used to add the BeautifulSoup folder the import path
import urllib2 # Used to read the html document

q_ = '永和'
url = 'http://www.google.com/search?q={}&start='.format(q_) +'&as_sitesearch=facebook.com'

if __name__ == "__main__":
    ### Import Beautiful Soup
    ### Here, I have the BeautifulSoup folder in the level of this Python script
    ### So I need to tell Python where to look.
    sys.path.append("./BeautifulSoup")
    from BeautifulSoup import BeautifulSoup

    ### Create opener with Google-friendly user agent
    opener = urllib2.build_opener()
    opener.addheaders = [('User-agent', 'Mozilla/5.0')]

    ### Open page & generate soup
    ### the "start" variable will be used to iterate through 10 pages.
    for start in range(0,50):
        #url = "http://www.google.com/search?q={}&start=".format(q_) + str(start*10)
        url = 'http://www.google.com/search?q={}&start={}'.format(q_,str(start*10)) +'&as_sitesearch=facebook.com' 
        #url = "http://www.google.com/search?q=taipei&start=" + str(start*10)
        print  url
        page = opener.open(url)
        soup = BeautifulSoup(page)
        for k in soup.findAll('h3'):
            if "<b>" in str(k.find('a').contents[0]):
                output[0].append(k.find('a').contents[1])
            else:
                output[0].append(k.find('a').contents[0])
        for k in soup.findAll('cite'):
            output[1].append(k.contents[0])
```

```python
pd.DataFrame(output).T#.to_csv('fb_google.csv')
```
